```python
import numpy as np
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0

E = 329.63
F = 349.23
G = 392.00
Ab = 415.30
A = 440.00
Bb = 466.16
C = 523.25
D = 587.33	
```

```python
# Carmen - Overture
notes=[C, D, 0, C, A, 0, A, 0, \
       A, 0, G, A, 0, Bb, A, 0, \
       Bb, 0, G, 0, C, A, 0, \
       F, 0, D/2, 0, G, C/2, 0, \
       G, 0, G, D, C, Bb, A, Ab, A, Bb, A, 0, \
       E, A, 0, A, 0, Ab, A, E*2] 
       
durs=[0.5, 0.25, 0.125, 0.125, 0.5, 0.125/2, 0.5, 0.125/2, \
      0.25, 0.125/2, 0.125, 0.25, 0.125/2, 0.125, 0.75, 0.25, \
      0.375, 0.125, 0.25, 0.125, 0.125, 0.625, 0.25,  \
      0.375, 0.125, 0.25, 0.125, 0.125, 0.625, 0.25, \
      0.625, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.25, 0.625, 0.25, \
      0.5, 0.5, 0.125/2, 0.5, 0.125/2, 0.375, 0.125, 1]    
      
      
      
      
      
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
